###  **Task 1: What is prompting?**
Prompting is the act of giving instructions, questions, or examples to a generative AI tool to guide it in producing a desired response or output.

### **Task 3: When you are using a gen AI tool to create a healthy three-course menu that requires a certain number of calories, which prompting element will ensure the menu is more practical?**
Adding context or constraints specifically, stating the required number of calories will make the menu more practical.

### **Task 4: Prompting with audio to receive a visual representation is called what?**
Multimodal prompting

### **Task 5: What does a human-in-the-loop approach to prompting emphasize?**
It emphasizes having a person review and verify AI outputs to ensure accuracy and relevance. Basically putting a human to ensure that the requirements are being satisfied.

### **Task 6: When the initial prompt "Create a dataset" is too broad, what should you do to refine it?**
I would make the prompt more specific by providing additional details, context, or maybe constraints about the dataset like topic, dataset size, structure, and use.

Despite its benefits, the internet comes with challenges. Cybersecurity threats, misinformation, and privacy concerns are ongoing issues.
