### **Task 1: Reflection Questions**

1) How is prompt chaining similar to solving a jigsaw puzzle?
Prompt chaining breaks a big problem into many small, connected pieces. Each prompt produces one piece. You then connect those pieces and the output of one becomes the input for the next, until the full picture emerges. Like a puzzle, small pieces are easier to place and verify than trying to assemble the whole at once.

2) What‚Äôs the main difference between iteration and chaining?
Iteration = keep rewording or tweaking a single prompt to get a better single response.
Chaining = run a sequence of targeted prompts where each builds on the previous output to accomplish a multi-step task.

3) Why might chaining be more effective than asking AI one big question?
Because large, complex tasks contain multiple subproblems (search, filter, synthesize, validate). A single prompt can overload the model or produce unfocused output. Chaining reduces cognitive load per step, preserves structure, makes errors easier to catch, and gives you checkpoints to verify before proceeding.

4) How can the context window of an AI model affect your chaining strategy?
Models can ‚Äúremember‚Äù only so much. If your chain produces long outputs, you must either summarize intermediate outputs before passing them on, or store large outputs externally and pass links/keys, or chunk the chain so each step stays within the context window.
